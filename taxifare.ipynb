```python
data['year']=data['date_time_of_pickup'].dt.year
data['month']=data['date_time_of_pickup'].dt.month
data['day']=data['date_time_of_pickup'].dt.day
data['hour']=data['date_time_of_pickup'].dt.hour
data['minute']=data['date_time_of_pickup'].dt.minute
data['second']=data['date_time_of_pickup'].dt.second
```

```python
def distance(lat1,lon1,lat2,lon2):
  p=0.017453292519943295
  a=0.5-np.cos((lat2-lat1)*p)/2+np.cos(lat1*p)*np.cos(lat2*p)*(1-np.cos((lon2-lon1)*p))/2
  return 0.6213712*12742*np.arcsin(np.sqrt(a))
data['distance']=distance(data.longitude_of_pickup, data.latitude_of_pickup,data.longitude_of_dropoff,data.latitude_of_dropoff)
```

```python
X = data.drop(columns=['unique_id','amount','date_time_of_pickup'])
y = data['amount']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
import sklearn
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
from sklearn.metrics import mean_squared_error
print("test rmse: %.3f"%mean_squared_error(y_test,y_pred)**0.5)
```
